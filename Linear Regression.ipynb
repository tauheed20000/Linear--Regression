## for simple linear regression we deal with on independent feature and one dependent feature
1. In this data we have 4 independent features and 1 dependent feature as sales.
2. we can combine three features as a independent features.
3. Unnamed feature should be remove.

```python
data.drop(columns=['Unnamed: 0'])
```

## combine into one independent feature as total_spend ]

```python
data['total_spend']=data['TV']+data['Radio']+data['Newspaper']
```

```python
data.head()
```

```python
data.drop(columns=['Unnamed: 0'],axis=1)
```

## find the relationship between sales and total_spend

```python
sns.scatterplot(data=data,x='total_spend',y='Sales')
```

## This clearly shows a linearity as more spend of advertising increase more sales    Now Ii want to draw a regresion line in abov.e 

```python
sns.regplot(data=data,x='total_spend',y='Sales',color='blue')
```
